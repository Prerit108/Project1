## Numerical and categorical feature


```python
cat_f = [feature for feature in df.columns if df[feature].dtype == "O"]
num_f = [feature for feature in df.columns if df[feature].dtype != "O"]
```

```python
print("Numerical feature :" , cat_f)
print("categorical features: ",num_f)
```

### Adding total score and average score

```python
df["total_score"] = df['math_score'] + df['reading_score'] + df['writing_score']
df["average_score"] = df["total_score"]/3
```

## Visulization

```python
fig,axs = plt.subplots(1,2,figsize = (16,7))
plt.subplot(1,2,1)
sns.histplot(data = df,x = df["average_score"],kde = True)
plt.subplot(1,2,2)
sns.histplot(data = df,x = df["average_score"],kde = True,hue = "gender")
```

```python
fig,axs = plt.subplots(1,2,figsize = (16,7))
plt.subplot(1,2,1)
sns.histplot(data = df,x = df["total_score"],kde = True)
plt.subplot(1,2,2)
sns.histplot(data = df,x = df["total_score"],kde = True,hue = "gender")
```
